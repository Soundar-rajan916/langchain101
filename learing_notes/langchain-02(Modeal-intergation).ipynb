### GROQ MODEL 

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY") 
```

```python
import os 
from langchain_groq import ChatGroq

model = ChatGroq(model="openai/gpt-oss-120b")
response = model.invoke("who is Hermione")
response.content
```

```python
for chunks in model.stream("explain the lore of one piece"):
    print(chunks.text ,end="", flush=True)
```

```python
responses = model.batch([
    "who is the villain in one piece anime",
    "what is the namde of the ship of strawhats in one piece",
    "who is luffy's father"
    ],
    config={
        'max-concurrency' :5
    }
)
for response in responses:
    print(response)
```
